```python
import pandas as pd
import folium
```

# Read Colombian cities locations

```python
locations = pd.read_csv('colombia.csv')
```

```python
locations.sample(5)
```

# Filter the department capital cities 

```python
capital_cities = locations[(locations.capital == 'admin') | (locations.capital == 'primary')]
```

```python
capital_cities.sample(5)
```

# Obtain the special district row

```python
special_district = capital_cities[capital_cities.capital == 'primary'].iloc[[0]]
```

```python
special_district
```

```python
map = folium.Map(location = (special_district.lat, special_district.lng), zoom_start = 6) 
```

```python
map
```

# Marks the points for each capital city

```python
popup_template = '<ul><li><b>Department:</b> {department}</li><li><b>City:</b> {city}</li><li><b>Population:</b> {population:,.0f}</li></ul>'
```

```python
for i, city in capital_cities.iterrows():
    folium.Marker([city.lat, city.lng], popup = popup_template.format(department = city.admin_name, city = city.city, population = city.population)).add_to(map)
```
